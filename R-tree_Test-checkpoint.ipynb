```python
import numpy as np
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import Point, MultiPoint
from shapely.geometry import Point, LineString
from shapely.ops import nearest_points
from shapely.geometry import Point
from shapely.geometry import Polygon
```

```python
# 创建经纬度数据
np.random.seed(0)  # 保证每次运行产生相同的随机数据
latitudes = np.random.uniform(-90, 90, 80)
longitudes = np.random.uniform(-180, 180, 80)

# 创建点对象
geometry_points = [Point(xy) for xy in zip(longitudes, latitudes)]

# 创建Geodataframe
points_within_bounds = gpd.GeoDataFrame(geometry=geometry_points)

# 创建经纬度数据
np.random.seed(1)  # 保证每次运行产生相同的随机数据
latitudes_road = np.random.uniform(-90, 90, 60).reshape(30, 2)
longitudes_road = np.random.uniform(-180, 180, 60).reshape(30, 2)

# 创建线对象
geometry_lines = [LineString([Point(xy[0]), Point(xy[1])]) for xy in zip(zip(longitudes_road[:, 0], latitudes_road[:, 0]), zip(longitudes_road[:, 1], latitudes_road[:, 1]))]

# 创建id列
road_ids = range(1, 31)

# 创建Geodataframe
chi_road = gpd.GeoDataFrame({'id': road_ids, 'geometry': geometry_lines})
```

```python
chi_road.crs = "EPSG:4326"
points_within_bounds.crs = "EPSG:4326"
chi_road = chi_road.to_crs("EPSG:4326")
points_within_bounds = points_within_bounds.to_crs("EPSG:4326")
```

# 普通的邻近点分析

```python
# Calculate the centroids for each line and Create unary union from the chi_road centroid points
chi_road['centroid'] = chi_road.centroid
unary_union = chi_road['centroid'].unary_union

# create a function 
def nearest(row, geom_union, df1, df2, geom1_col='geometry', geom2_col='centroid', src_column=None):
    """Find the nearest point and return the corresponding value from specified column."""
    
    # Find the geometry that is closest
    nearest = df2[geom2_col] == nearest_points(row[geom1_col], geom_union)[1]
    
    # Get the corresponding value from df2 (matching is based on the geometry)
    value = df2[nearest][src_column].values[0]
    
    return value

# Store the 'id' of the nearest "chi_road" points into a new column "nearest_id" in "points_within_bounds"
points_within_bounds['nearest_id'] = points_within_bounds.apply(nearest, geom_union=unary_union, df1=points_within_bounds, df2=chi_road, geom1_col='geometry', src_column='id', axis=1)
points_within_bounds
```

```python
# 执行合并操作
Chiroad_Police = pd.merge(points_within_bounds, chi_road, left_on='nearest_id', right_on='id')
Chiroad_Police = Chiroad_Police.rename(columns={'geometry_x':'geometry_policestop','geometry_y':'geometry_roads'})

# Assuming that 'geometry_points' and 'geometry_lines' are your column names
Chiroad_Police_points = gpd.GeoDataFrame(Chiroad_Police, geometry='geometry_policestop')
Chiroad_Police_lines = gpd.GeoDataFrame(Chiroad_Police, geometry='geometry_roads')

Chiroad_Police['nearest_id'].unique()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,1, figsize=(20,16))

# plot lines
Chiroad_Police_lines.plot(ax=ax, column='id', cmap='viridis', linewidth=2)

# plot points
Chiroad_Police_points.plot(ax=ax, column='id', marker='o', cmap='viridis', markersize=20)

plt.show()
```

# R-tree邻近点分析

```python
import geopandas as gpd
from shapely.geometry import Point
from shapely.ops import nearest_points
from rtree import index

# Calculate the centroids for each line
unary_union = chi_road.unary_union

# Prepare R-tree
idx = index.Index()
for i, geom in enumerate(chi_road.geometry):
    idx.insert(i, geom.bounds)

# Create a function
def nearest(row, idx, df1, df2, geom1_col='geometry', src_column=None):
    """Find the nearest line and return the corresponding value from specified column."""
    # Find the nearest points (using the R-tree index)
    nearest_geom_ids = list(idx.nearest((row[geom1_col].x, row[geom1_col].y), 1))
    
    # Initialize the min_distance with a large number
    min_distance = float("inf")
    min_id = None
    for nearest_id in nearest_geom_ids:
        nearest_line = df2.loc[nearest_id].geometry
        distance = nearest_line.distance(Point(row[geom1_col].x, row[geom1_col].y))
        if distance < min_distance:
            min_distance = distance
            min_id = nearest_id

    return df2.loc[min_id][src_column]

# Store the 'id' of the nearest "chi_road" points into a new column "nearest_id" in "points_within_bounds"
points_within_bounds['nearest_id'] = points_within_bounds.apply(nearest, idx=idx, df1=points_within_bounds, df2=chi_road, geom1_col='geometry', src_column='id', axis=1)
points_within_bounds
```

```python
# 执行合并操作
Chiroad_Police = pd.merge(points_within_bounds, chi_road, left_on='nearest_id', right_on='id')
Chiroad_Police = Chiroad_Police.rename(columns={'geometry_x':'geometry_policestop','geometry_y':'geometry_roads'})

# Assuming that 'geometry_points' and 'geometry_lines' are your column names
Chiroad_Police_points = gpd.GeoDataFrame(Chiroad_Police, geometry='geometry_policestop')
Chiroad_Police_lines = gpd.GeoDataFrame(Chiroad_Police, geometry='geometry_roads')

Chiroad_Police['nearest_id'].unique()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,1, figsize=(20,16))

# plot lines
Chiroad_Police_lines.plot(ax=ax, column='id', cmap='viridis', linewidth=2)

# plot points
Chiroad_Police_points.plot(ax=ax, column='id', marker='o', cmap='viridis', markersize=20)

plt.show()
```
