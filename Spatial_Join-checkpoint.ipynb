## Use the road network as a bound to filter out outlier points

```python
chi_road.bounds
```

```python
min_lat = chi_road.bounds.miny.min()
min_lon = chi_road.bounds.minx.min()
max_lat = chi_road.bounds.maxy.max()
max_lon = chi_road.bounds.maxx.max()
print(min_lat)
print(min_lon)
print(max_lat)
print(max_lon)
```

<font color='EE8179'>
    
- I got the Latitude and Longitude bounds for Chicago_road

```python
from shapely.geometry import Polygon

# Create a boundary
Top_left_corner = (min_lon, max_lat)
Top_right_corner = (max_lon, max_lat)
Lower_right_corner = (max_lon, min_lat)
Lower_left_corner = (min_lon, min_lat)

# 创建一个Polygon对象来代表这个经纬度范围
bounds = Polygon([Top_left_corner, Top_right_corner, Lower_right_corner, Lower_left_corner])

# 使用within函数选择在这个经纬度范围内的所有点
points_within_bounds = police_stop[police_stop.geometry.within(bounds)]
```

```python
points_within_bounds = points_within_bounds.to_crs(chi_road.crs)
```

<font color='EE8179'>
    
- I use the bounds of Chicago_road to filter outlier "police_stop" points, and gave a new Geodataframe name for "police_stop" dataset, the new name is "points_within_bounds".

```python
# Filtered 4669 data that are no longer within the boundaries
print(len(police_stop)-len(points_within_bounds))
```

<font color='EE8179'>
    
- We can find there are about 4,669 points were filtered!

# Nearest Neighbour Analysis

[Following the code of the Nearest Neighbour Analysis website, it can be run, but it will take a lot of time because it is exhaustive approach, so we have to find more advanced methods](https://autogis-site.readthedocs.io/en/2018_/notebooks/L3/nearest-neighbour.html)

Using R-tree for spatial indexing:  
- Modify point-point distance to point-line distance  
- Enhanced operation speed with spatial indexing

```python
import geopandas as gpd
from shapely.geometry import Point
from shapely.ops import nearest_points
from rtree import index

# Calculate the centroids for each line
unary_union = chi_road.unary_union

# Prepare R-tree
idx = index.Index()
for i, geom in enumerate(chi_road.geometry):
    idx.insert(i, geom.bounds)

# Create a function
def nearest(row, idx, df1, df2, geom1_col='geometry', src_column=None):
    """Find the nearest line and return the corresponding value from specified column."""
    # Find the nearest points (using the R-tree index)
    nearest_geom_ids = list(idx.nearest((row[geom1_col].x, row[geom1_col].y), 1))
    
    # Initialize the min_distance with a large number
    min_distance = float("inf")
    min_id = None
    for nearest_id in nearest_geom_ids:
        nearest_line = df2.loc[nearest_id].geometry
        distance = nearest_line.distance(Point(row[geom1_col].x, row[geom1_col].y))
        if distance < min_distance:
            min_distance = distance
            min_id = nearest_id

    return df2.loc[min_id][src_column]

# Store the 'id' of the nearest "chi_road" points into a new column "nearest_id" in "points_within_bounds"
points_within_bounds['nearest_id'] = points_within_bounds.apply(nearest, idx=idx, df1=points_within_bounds, df2=chi_road, geom1_col='geometry', src_column='id', axis=1)
points_within_bounds
```

<font color='EE8179'>
    
- I use "R-tree" methods to merge "chicago_road['id']" into "points within bounds" file, which means for each "police_stop" point, give it the nearest road id.  
- Note: We have 647,188 data rows, but we only have 234,888 different roads id, which means for "points_within_bounds['nearest_id']", we have lots of repeated values. That make sense, because for each road, maybe there are many point close to it.

# Merge "chicago_road" file into "points_within_bounds" file by the only road id

<font color='EE8179'>
    
- I merge "chicago_road" file into "points_within_bounds" file by the only road_id and give it a new Geodatframe name "Chiroad_Police" to store data

```python
# 执行合并操作
Chiroad_Police = pd.merge(points_within_bounds, chi_road, left_on='nearest_id', right_on='id')
Chiroad_Police = Chiroad_Police.rename(columns={'geometry_x':'geometry_policestop','geometry_y':'geometry_roads'})
Chiroad_Police
```

```python
# Assuming that 'geometry_points' and 'geometry_lines' are your column names
Chiroad_Police_points = gpd.GeoDataFrame(Chiroad_Police, geometry='geometry_policestop')
Chiroad_Police_lines = gpd.GeoDataFrame(Chiroad_Police, geometry='geometry_roads')
```
